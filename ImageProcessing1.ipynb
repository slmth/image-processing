## BLENDING AND PASTING IMAGES

```python
img1=cv2.imread("2081.jpg")
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
img2=cv2.imread("773.jpg")
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
img1=cv2.resize(img1,(250,250))
img2=cv2.resize(img2,(250,250))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
blended=cv2.addWeighted(src1=img1,alpha=0.5,src2=img2,beta=0.5,gamma=0)
plt.imshow(blended)
#addWeighted working with only same sizes images
```

```python
img1=cv2.imread("ii.png")
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.imread("bg.jpg")
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img1=cv2.resize(img1,(200,200))
plt.imshow(img1)
```

```python
img2.shape
plt.imshow(img2)
```

```python
x_offset=500-200
y_offset=332-200
rows,cols,channels=img1.shape
```

```python
roi=img2[y_offset:332,x_offset:500]
plt.imshow(roi)
```

```python
img1gray=cv2.cvtColor(img1,cv2.COLOR_RGB2GRAY)
plt.imshow(img1gray,cmap="gray")
```

```python
mask_inv=cv2.bitwise_not(img1gray)
plt.imshow(mask_inv,cmap="gray")
```

```python
import numpy as np
white_bg=np.full(img1.shape,255,dtype=np.uint8)
white_bg.shape
```

```python
bk=cv2.bitwise_or(white_bg,white_bg,mask=mask_inv)
```

```python
plt.imshow(bk)
```

```python
fg=cv2.bitwise_or(img1,img1,mask=mask_inv)
plt.imshow(fg)
```

```python
final_roi=cv2.bitwise_or(roi,fg)
plt.imshow(final_roi)
```

```python
lim=img2
sim=final_roi
lim[y_offset:y_offset+sim.shape[0],x_offset:x_offset+sim.shape[1]]=sim
plt.imshow(lim)
```

```python
im2=cv2.imread("bulmaca.jpg")
plt.imshow(im2,cmap="gray")
```

```python
def show_pic(im2):
    fig=plt.figure(figsize=(15,15))
    ax=fig.add_subplot(111)
    ax.imshow(im2,cmap="gray")
```

```python
show_pic(im2)
```

```python
ret2,th2=cv2.threshold(im2,127,255,cv2.THRESH_BINARY)
show_pic(th2)
```
